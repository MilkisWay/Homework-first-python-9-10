```python
def NextStep(n,x,y,dx,dy):
    if x==0:
        if dx!=0 and dy!=0:
            # вышли на границу х=0
            if y<n-1:
                # скользим
                dx=0
                dy=1
            else:
                #мы в углу скользим по другой грани
                dx=1
                dy=0
        else:
        #уже скользнули по грани
            dx=1
            dy=-1
    elif x==n-1:
        if dx!=0 and dy!=0:
        # вышли на границу х=n-1
            if y<n-1:
                # скользим
                dx=0
                dy=1
            else:
                #мы пришли
                dx=0
                dy=0
        else:
        #уже скользнули по грани
            dx=-1
            dy=1
    elif y==0:
        if dx!=0 and dy!=0:
        # вышли на границу y=0
            if x<n-1:
                # скользим
                dx=1
                dy=0
            else:
                #мы в углу скользим по другой грани
                dx=0
                dy=1
        else:
        #уже скользнули по грани
            dx=-1
            dy=1
    elif y==n-1:
        if dx!=0 and dy!=0:
        # вышли на границу y=n-1
            if x<n-1:
                # скользим
                dx=1
                dy=0
            else:
                #мы пришли
                dx=0
                dy=0
        else:
        #уже скользнули по грани
            dx=1
            dy=-1
   
    return dx,dy

n=6 #задаем размер матрицы
Matrica=np.zeros((n,n))
Matrica[0,0]=0
Matrica[1,0]=1
Matrica[0,1]=2
dx=1
dy=-1
x=1
y=0

for k in range(3,n**2):
    dx,dy=NextStep(n,x,y,dx,dy)
    x=x+dx
    y=y+dy
    Matrica[y,x]=k
print(Matrica)
```

```python
import numpy as np


def snake(n,m):
    arr=np.zeros((n,m))#матрица размером n*m из нулей
    i,j,val = 0,0,0
    while val<n*m:#пока вся матрица не заполнена
        arr[i][j]=val #присвоить элементу значение
        i,j=i+1,j-1
        if i>n-1:
            arr=arr.T
            i,j=j+2,i-1
            n,m=m,n
        elif j<0:
            arr=arr.T
            i,j=0,i
            n,m=m,n
        val+=1
    return arr

try:
    n,m =map(int,input().split())
    print(snake(n,m)) 
except ValueError:
    print('Неправильный ввод')
```
